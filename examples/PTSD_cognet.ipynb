```python
from quasinet.qnet import qdistance
from cognet.cognet import cognet as cg
from cognet.dataFormatter import dataFormatter
from cognet.model import model 
#import cognet.util
import pandas as pd
import numpy as np
```

```python
## read in data, configure to Qnet specifications and fit Qnet model
fit = False
data_obj=dataFormatter("../../PTSD_cognet_test_processed.csv")
model_obj = model()
if fit:
    model_obj.fit(data_obj=data_obj)
    model_obj.save("../../mpi_tmp/PTSD_cognet_test.joblib")
else:
    model_obj.load("../../mpi_tmp/PTSD_cognet_test.joblib")
```

```python
# load Qnet, write mpi files for distance matrix, and compute dissonance matrix
cognet_obj = cg()
cognet_obj.load_from_model(model_obj, data_obj, 'all')
cognet_obj.dmat_filewriter("examples_results/PTSD_cognet.py", "examples_results/PTSD_cognet_test.joblib",NODES=4,T=3)
cognet_obj.samples
```

```python
return_dict = cognet_obj.dissonance_matrix(output_file="mpi_tmp/PTSD_dissonance_matrix.csv", n_jobs=4)
return_dict
```

```python
# embed distance matrix
cognet_obj.year = '2016'
cognet_obj.embed("../../mpi_tmp/distmatrix.csv", "PTSD", "mpi_tmp/",EMBED_BINARY='cognet/cognet/bin/__embed__.so')
```

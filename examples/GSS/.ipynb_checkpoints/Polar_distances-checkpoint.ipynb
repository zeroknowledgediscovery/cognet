```python
# This is an example of how we see 
# the package work. The functions listed here
# are probably the only ones that should be exposed, ie documented.
# others should br prepended with a double underscore
#  
# The cognet directory has the following "modules"
# which are seprate .py files containing clases and functions
# The modules are cognet.py, dataFormatter.py, model.py, util.py, viz.py
# we will write the viz.py later.
import sys

from quasinet.qnet import qdistance
from cognet.cognet import cognet as cg
from cognet.dataFormatter import dataFormatter
from cognet.model import model 
#import cognet.util
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

yr = '2018'
POLEFILE='data/polar_vectors.csv'
QPATH='data/gss_'+yr+'.joblib'
IMMUTABLE_FILE='data/immutable.csv'
GSSDATA = 'data/gss_'+yr+'.csv'
```

```python
# testing dataFormatter
data = dataFormatter(samples=GSSDATA)
# load the sample data
# have option for test/train split
# make checks to ensure we will not throw errors at qnet construction 
print(data.samples[:2])
features,samples = data.format_samples('train') # default trains and tests using half
all_samples = True
if all_samples: # use all samples to train, instead of half
    features,samples = data.Qnet_formatter()

# testing model functionality
# can either input features and samples directly, or infer from data obj
model_ = model()

# qnet construction parameters, 
# choose to either load or fit qnet from scratch
# and to either load from url or local repo
test_model_buildqnet = False
url_load = True
if url_load:
    QNETFILE = 'https://zenodo.org/record/5781768/files/gss_2018.joblib'
else:
    QNETFILE = 'data/gss_2018.joblib'
model_.load(QNETFILE)

# testing cognet
# set some paramaters in instantiating cognet class 
# if loading from model obj, no need to use load_data func, otherwise, load_data
Cg = cg()
print(len(model_.features))
Cg.load_from_model(model_, data, 'all')
```

```python
Cg.set_poles(POLEFILE,"L","R", steps=25, VERBOSE=True, restrict=True)

# compute polar distance matrix
Cg.polar_separation(nsteps=0)[1][0]
```

```python
# calculate polar separation as qsample steps increase
pole_distance = pd.DataFrame()
ntimes = 10

# read poles
poles = pd.read_csv(POLEFILE, index_col=0)
poles=poles.transpose()
polar_features = pd.concat([Cg.features, Cg.poles], axis=0).fillna('')

# initial pole and base distance
pR = polar_features.loc["R"][Cg.cols].fillna('').values.astype(str)[:]
pL = polar_features.loc["L"][Cg.cols].fillna('').values.astype(str)[:]
d0 = qdistance(pR, pL, Cg.qnet, Cg.qnet)

variation_means = []
R_ideology_means = []
L_ideology_means = []

for i in range(1, 80):
    i = i * 10
    print(i)
    variations = []
    R_ideology = []
    L_ideology = []
    
    # take 10 qsamples
    for j in range(0, 10):
        pR_ = Cg.qsampling(pR,i)
        pL_ = Cg.qsampling(pL,i)
        
        RdR = qdistance(pR_, pR, Cg.qnet, Cg.qnet)
        RdL = qdistance(pR_, pL, Cg.qnet, Cg.qnet)
        
        LdR = qdistance(pL_, pR, Cg.qnet, Cg.qnet)
        LdL = qdistance(pL_, pL, Cg.qnet, Cg.qnet)
        
        Rideology_index = (RdR-RdL)/d0
        Lideology_index = (LdR-LdL)/d0
        
        R_ideology.append(Rideology_index)
        L_ideology.append(Lideology_index)
        variations.append(qdistance(pR_, pL_, Cg.qnet, Cg.qnet))
    
    # calculate mean for final df, can also make each column nsteps, and rows mean values
    variation_means.append(sum(variations) / 10)
    R_ideology_means.append(sum(R_ideology) / 10)
    L_ideology_means.append(sum(L_ideology) / 10)
pole_distance["distance"] = variation_means
pole_distance["R_ideology"] = R_ideology_means
pole_distance["L_ideology"] = L_ideology_means
```

```python
pole_distance["variation"] = variation_means
pole_distance["R_ideology"] = R_ideology_means
pole_distance["L_ideology"] = L_ideology_means
```

```python
pole_distance["nsteps"] = [i * 10 for i in range(1, 80)]
pole_distance.to_csv("results/polar_distances_nsteps.csv")
pole_distance
```

```python
#plt.gca().set_color_cycle(['red', 'blue'])
labels = []
fig, ax = plt.subplots()
ax.plot(pole_distance["nsteps"], pole_distance["variation"], c="green")
labels.append(r"Polar Separation")
ax.set_xlabel("Number of Steps")
ax.set_ylabel("Qdistance between Poles")
plt.legend(labels, loc="upper left")

ax2 = ax.twinx()

ax2.plot(pole_distance["nsteps"], pole_distance[["R_ideology", "L_ideology"]])
ax2.set_ylabel("Ideology Index")
plt.title("Polar Separation and Ideology Index")
colors = ["red","blue"]
labelnames = ["R", "L"]
labels = []
for i,j in enumerate(ax2.lines):
    j.set_color(colors[i])
    labels.append(r"{} Ideology".format(labelnames[i]))
    
plt.legend(labels)
plt.show()


# pole_distance.plot()
# plt.xlabel("nsteps")
# plt.ylabel("Qdistance")
# plt.title("distance between poles")
# plt.savefig("results/")
```
